## 20. How would you write a regex that matche a number with comma for every three digits? It must match the given following:

&#39;42&#39;

&#39;1,234&#39;

&#39;6,368,745&#39;

but not the following:


&#39;12,34,567&#39; (which has only two digits between the commas)

&#39;1234&#39; (which lacks commas)

```python
import re

s = '42 1,234 6,368,745 12,34,567 1234'

rgx = re.compile(',[0-9]{1,2},|[0-9]{4,}')
num = [x for x in s.split() if not rgx.search(x)]

print(num)  #['42', '1,234', '6,368,745']
```

## 21. How would you write a regex that matches the full name of someone whose last name is Watanabe? You can assume that the first name that comes before it will always be one word that begins with a capital letter. The regex must match the following:

&#39;Haruto Watanabe&#39;

&#39;Alice Watanabe&#39;

&#39;RoboCop Watanabe&#39;

but not the following:

&#39;haruto Watanabe&#39; (where the first name is not capitalized)

&#39;Mr. Watanabe&#39; (where the preceding word has a nonletter character)

&#39;Watanabe&#39; (which has no first name)

&#39;Haruto watanabe&#39; (where Watanabe is not capitalized)

```python
import re

def printMatch(name):
    pat = re.compile(r'\b[A-Z][a-zA-Z]+\sHaruto Watanabe')
    if pat.search(name):
        print('"'+name+'" matches')
    else:
        print('"'+name+'" does not match')

printMatch('test Haruto Watanabe test')
printMatch('test Alice Watanabe test')
printMatch('test RoboCop Nakamoto test')
```
