## 2. What is the value of a dictionary value with the key &#39;foo&#39; and the value 42?

* {'foo': 42},Here the 'foo' is key and 42 is a value in dict.

Ex-- which show the Key and remaining one is value.

```python
# creating a new dictionary
my_dict ={"foo":42, "python":112, "c":11}
 
# list out keys and values separately
key_list = list(my_dict.keys())
val_list = list(my_dict.values())
 
# print key with val 100
position = val_list.index(42)
print(key_list[position])
 
# print key with val 112
position = val_list.index(112)
print(key_list[position])
 
```
